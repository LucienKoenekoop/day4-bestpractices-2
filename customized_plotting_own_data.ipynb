## 2. The default plot

Let's create some data (normal distribution)

````python
import numpy as np
import pandas as pd
data = pd.read_csv ('own_data.csv')
data
````

and plot a simple histogram of our data

````python
fig = plt.figure(figsize=(16,9))
x = data['min_per_match']
y1 = data['rank_in_league_top_attackers']
y2 = data['rank_in_league_top_midfielders']
y3 = data['rank_in_league_top_defenders']
# x, y
plt.scatter(x, y1, marker='^', c='b')
plt.scatter(x, y2, marker='o', c='g')
plt.scatter(x, y1, marker='*', c='y')
plt.xlabel('minutes per match')
plt.ylabel('rank')
plt.plot()
plt.show()
````
